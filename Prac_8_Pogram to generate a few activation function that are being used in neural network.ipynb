```python
#Binary Sigmoid Function:
The binary sigmoid function squashes input values between 0 and 1, commonly used in binary classification problems. It smoothly transforms any real number to a probability.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Function
plt.style.use('seaborn')
plt.figure(figsize=(8, 4))

def sigmoid_binary(t):
    return 1 / (1 + np.exp(-t))

t_sigmoid = np.linspace(-5, 5)
plt.plot(t_sigmoid, sigmoid_binary(t_sigmoid))
plt.title('Binary Sigmoid Activation Function')
plt.show()
```

```python
#Hyperbolic Tangent Function (tanh):
The hyperbolic tangent function, tanh, squashes input values between -1 and 1, facilitating centered data. It is commonly used in hidden layers of neural networks.
```

```python
# Hyperbolic Tangent Function
plt.style.use('seaborn')
plt.figure(figsize=(8, 4))

def hyperbolic_tan(t):
    return np.tanh(t)

t_tanh = np.linspace(-5, 5)
plt.plot(t_tanh, hyperbolic_tan(t_tanh))
plt.title('Hyperbolic Tangent Activation Function')
plt.show()
```

```python
#Rectified Linear Unit (ReLU):
ReLU activation sets negative input values to zero and leaves positive values unchanged. It is widely used in neural networks due to its simplicity and effectiveness in mitigating the vanishing gradient problem.
```

```python
# Rectified Linear Unit (ReLU)
plt.style.use('seaborn')
plt.figure(figsize=(8, 4))

def rectified_linear_unit(t):
    return np.maximum(0, t)

t_relu = np.linspace(-5, 5)
plt.plot(t_relu, rectified_linear_unit(t_relu))
plt.title('Rectified Linear Unit (ReLU) Activation Function')
plt.show()
```

```python
#Binary Step Function:
The binary step function assigns 1 for non-negative inputs and 0 for negative inputs. It serves as a basic thresholding function, often used in simple perceptrons.
```

```python
# Binary Step Function
plt.style.use('seaborn')
plt.figure(figsize=(8, 4))

def binary_step(x):
    return np.where(x >= 0, 1, 0)

x_binary_step = np.linspace(-10, 10)
plt.plot(x_binary_step, binary_step(x_binary_step))
plt.axis('tight')
plt.title('Activation Function: Binary Step')
plt.show()
```

```python
#Linear Activation Function:
The linear activation function preserves the input as the output, commonly used in regression tasks. It allows the network to learn a direct relationship between input and output.
```

```python
# Linear Activation Function
plt.figure(figsize=(8, 4))

def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x

x_linear = np.linspace(-10, 10)
plt.plot(x_linear, linear(x_linear))
plt.axis('tight')
plt.title('Activation Function: Linear')
plt.show()
```

```python
#Softmax Activation Function:
The softmax function converts a vector of real numbers into a probability distribution, often used in the output layer of a neural network for multi-class classification tasks. It normalizes the input to sum to 1, making it suitable for probability interpretation.
```

```python
# Softmax Activation Function
plt.style.use('seaborn')
plt.figure(figsize=(8, 4))

def softmax(t):
    exp_t = np.exp(t)
    return exp_t / np.sum(exp_t)

t_softmax = np.linspace(-5, 5)
plt.plot(t_softmax, softmax(t_softmax))
plt.title('Softmax Activation Function')
plt.show()
```
